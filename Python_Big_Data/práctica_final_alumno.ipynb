```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
def total_price(bought,list_of_things):
    total_p = 0
    for c in bought:
        total_p += list_of_things[c]
    print(f"El precio total de la compra es: {total_p}")

total_price(compra, articulos)

# B. Una función que calcule el precio total de la compra con un descuento del 10%
def discount_10(bought,list_of_things):
    total_p = 0
    for c in bought:
        total_p += list_of_things[c]
        disc = total_p * 0.10
        new_total = total_p - disc
    print(f"El precio total de la compra con un descuento del 10 % es: {new_total}")

discount_10(compra, articulos)


# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def discount_if_more_100(bought, list_of_things):
    total_p = 0
    for c in bought:
        total_p += list_of_things[c]
    if total_p >= 100:
        disc = total_p * 0.10
        new_total = total_p - disc
        print(f"El precio total de la compra con un descuento del 10% porqué la compra supera los 100€: {new_total}")
    else:
        print(f"El precio total de la compra sin descuento: {total_p}")

discount_if_more_100(compra, articulos)

# D. Una función que calcule el precio total aplicando el IVA (21%)

def total_price_vat(bought, list_of_things):
    total_p = 0
    for c in bought:
        total_p += list_of_things[c]
    vat = total_p * 0.21
    new_total = total_p + vat
    print(f"El precio total aplicando el 21% de IVA es: {new_total}")

total_price_vat(compra, articulos)

# E.lista los artículos cuyo precio es superior a 20€

superior_20 = {new_key:new_value for (new_key, new_value) in articulos.items() if new_value > 20}
print(f"Los articulos cuyo precio es superior a 20 € son: {superior_20}")

```
